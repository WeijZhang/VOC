```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
biomass=pd.read_csv('All.csv')
X = biomass.values[:,0:12]#定义x
Y = biomass.values[:,12:15]#定义单目标y
biomass.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=110)
from sklearn.multioutput import MultiOutputRegressor
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.ensemble import GradientBoostingRegressor
#n_estimators=range(19,31,1)
n_estimators=[4,8,16,32,64,100,128,150]
max_depth=[1,2,3,4,8,16,32,64,100,128]
#max_depth=range(1,11,1)
learning_rate=[0.0001,0.001,0.01,0.1,1]
R2=[]
R_STD=[]

for i in n_estimators:
    for j in max_depth:
        model =MultiOutputRegressor(GradientBoostingRegressor(learning_rate=0.1, n_estimators=i, loss= 'ls',max_depth=j,random_state=1))
        score=cross_val_score(model, X_train, y_train, cv=5)#scoring="neg_mean_squared_error"
        R2.append(score.mean())
        R_STD.append(score.std())
        print(i,j,score.mean())
#print(n_estimators[R2.index(max(R2))],max_depth[R2.index(max(R2))],max(R2))
print(max(R2))
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error

model =MultiOutputRegressor(GradientBoostingRegressor(loss= 'ls',learning_rate=0.1, n_estimators=128, max_depth=3,random_state=1))
model.fit(X_train,y_train)
y_train_pre=model.predict(X_train)
y_test_pre=model.predict(X_test)
print(model.score(X_test,y_test))
r=r2_score(y_train,y_train_pre,multioutput='raw_values')
r2=r2_score(y_test, y_test_pre,multioutput='raw_values')
r3=mean_absolute_error(y_train,y_train_pre,multioutput='raw_values')
r4=mean_absolute_error(y_test, y_test_pre,multioutput='raw_values')
tMSE=mean_squared_error(y_train, y_train_pre,multioutput='raw_values')
tMSE=tMSE**0.5
MSE=mean_squared_error(y_test, y_test_pre,multioutput='raw_values')
MSE=MSE**0.5

print('train R2',np.around(r,2),'average train R2:',np.around(np.mean(r,axis=0),2),'test R2:',np.around(r2,2),'average test R2:',np.around(np.mean(r2,axis=0),2),'train RMSE:',np.around(tMSE,2), 'average train RMSE:',np.around(np.mean(tMSE,axis=0),2),'test RMSE:',np.around(MSE,2),'average test RMSE:',np.around(np.mean(MSE,axis=0),2),
    'train MAE:',np.around(r3,2), 'average train MAE:',np.around(np.mean(r3,axis=0),2),'test MAE:',np.around(r4,2),'average test MAE:',np.around(np.mean(r4,axis=0),2))
```
